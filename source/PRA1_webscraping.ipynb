```python
#Importar librer√≠as requeridas
import requests
from bs4 import BeautifulSoup as bs
import random
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import undetected_chromedriver as uc
```

```python
#Se define Google Chrome el navegador a utilizar 
browser = uc.Chrome()
```

```python
# En este bloque se define un √°rea espec√≠fica de inter√©s para la b√∫squeda de inmuebles,
# el script ingresa a la p√°gina https://www.fincaraiz.com.co simulando dicha b√∫squeda, 
# recorre todas las p√°ginas de resultado y recopila las url de todos los inmueble que
# cumplen con la condici√≥n especif√≠cada, generando una lista de url de referencia 

busqueda = 'Engativa' # Barrio, localidad o Ciudad de la cu√°l se desea extraer informaci√≥n
# Para este ejercicio se hace la b√∫squeda para la localidad de Engativ√°, que es una de las
# m√°s grandes en la ciudad de Bogot√°
busqueda = busqueda.replace(' ','-')
x=1
refs = []

while True:   
    url = f'https://www.fincaraiz.com.co/apartaestudios-apartamentos-casas/venta?usado=true&ubicacion={busqueda}&pagina={x}'
    browser.get(url)
    # Se define un tiempo de retardo entre b√∫squeda con el fin de no saturar el servidor
    time.sleep(random.randint(10,12))
    html = browser.page_source
    soup = bs(html, 'html.parser')
    # Esta secci√≥n permite navegar entre la totalidad de las p√°ginas de resultado con el fin de extraer todas
    # url de los inmuebles que cumplen la condici√≥n inicial, al llegar a la √∫ltima pa«µina el script termina
    try:        
        pagina_actual = int(soup.
                            find('ul',{'class':'MuiPagination-ul'}).
                            find('button',
                                 {'class':'MuiButtonBase-root MuiPaginationItem-root MuiPaginationItem-page MuiPaginationItem-outlined MuiPaginationItem-rounded Mui-selected'}).
                            get('aria-label').
                            replace('page ',''))
    except AttributeError:
        pagina_actual = 1
    if x == pagina_actual:
        articles = soup.find('div',{'id':'listingContainer'}).find_all('a')
        # el id listingContainer contiene las url de referencia
    else:
        break    
    x = x+1
    #------------------------------------------------------------------------------
    for article in articles:
        href = article.get('href')
        refs.append(href)
        time.sleep(random.randint(1,3))
        print(href)
    
    #Se crea una lista con las url de referencia
    refs = [href for href in refs if href is not None]
    
print ('¬°Proceso Terminado!')
```

```python
# Se crea dataFrame con url de referencia
href_inmueble = pd.DataFrame(refs)
href_inmueble.columns = ['href']
href_inmueble.shape
```

```python
# Se exporta dataFrame de url a archivo CSV como copia de respaldo
href_inmueble.to_csv('referencias_inmuebles.csv', index = False)
```

```python
# Se crea un objeto de tipo Series que servir√° como repositorio para los datos obtenidos de cada inmueble
inmuebles = pd.Series()
```

```python
# Se crea funci√≥n encargada de recorrer cada url de referencia, extraer los datos m√°s relevantes:
# Esta funci√≥n recorre p√°gina a p√°gina, extrae la informaci√≥n contenida en los elmentos div con id
# general (aqu√≠ se encuentran los datos m√°s relevantes del inmuble; habitaciones, ba√±os, parqueadero,
# etc..) m√°s los elementos del elemento div con id location (Datos de ubicaci√≥n geogr√°fica del inmueble),
# estos datos son limpiados y almacenados en un diccionario de donde se tomar√°n los datos finales


def extraeInfoInmueble(href_inmueble):


    print('\n Inmueble referenciado: ' + href_inmueble)
    url = "https://www.fincaraiz.com.co" + href_inmueble + "/"
    browser.get(url)
    html = browser.page_source
    soup = bs(html, 'html.parser')
    try:
        general = soup.find('div', {'id': 'general'})
        caracteristicas = general.findAll(attrs={'class': 'jss75'})
        atributos = caracteristicas[2:-1]
        location = soup.find('div',{'id':'location'})
        datosUbicacion=location.findAll(attrs={'class':'jss74'})
        dic_inmueble = dict()
        dic_inmueble['codigoInmueble'] = caracteristicas[0].text.split(':')[1].strip()
        for x in range(0, len(atributos), 2):
            dic_inmueble[atributos[x].text] = atributos[(x + 1)].text

        for x in range(0, len(datosUbicacion),2):
            dic_inmueble[datosUbicacion[x].text] = datosUbicacion[(x+1)].text

        titulo = soup.find('p', {'class': 'jss66 jss71 jss125'}).text
        codigoInmueble = dic_inmueble.get('codigoInmueble', 'Sin informaci√≥n')
        areaPrivada = dic_inmueble.get('√Årea privada', 'Sin informaci√≥n')
        areaConstruida = float(dic_inmueble.get('√Årea constru√≠da', '0').
                               replace('.', '').replace(',', '.').
                               replace('m¬≤', '').strip())
        precioM2 = int(dic_inmueble.get('Precio m¬≤', '0').replace('.','').
                       replace('$\xa0','').replace('*m¬≤','').split(',')[0])
        estrato = dic_inmueble.get('Estrato', 'Sin informaci√≥n')
        antiguedad = dic_inmueble.get('Antig√ºedad', 'Sin informaci√≥n')
        habitaciones = int(dic_inmueble.get('Habitaciones', '0'))
        banios = int(dic_inmueble.get('Ba√±os', '0'))
        parqueaderos = int(dic_inmueble.get('Parqueaderos', '0'))
        barrio = dic_inmueble.get('Barrio', 'Sin informaci√≥n')
        direccion = dic_inmueble.get('Direcci√≥n', 'Sin informaci√≥n')
        precio = int(areaConstruida * precioM2)

        inmuebles['TITULO'] = titulo
        inmuebles['CODIGO INMUEBLE'] = codigoInmueble
        inmuebles['√ÅREA PRIVADA'] = areaPrivada
        inmuebles['√ÅREA CONSTRU√çDA'] = str(areaConstruida) + 'm¬≤'
        inmuebles['PRECIO POR M¬≤'] = precioM2
        inmuebles['ESTRATO'] = estrato
        inmuebles['ANTIGUEDAD'] = antiguedad
        inmuebles['HABITACIONES'] = habitaciones
        inmuebles['BA√ëOS'] = banios
        inmuebles['PARQUEADEROS'] = parqueaderos
        inmuebles['BARRIO'] = barrio
        inmuebles['DIRECCI√ìN'] = direccion    
        inmuebles['PRECIO'] = precio

        df_inmuebles = pd.DataFrame(inmuebles)
        
        return (df_inmuebles.T)
    
    except AttributeError:
        print (f'No se encuentra informaci√≥n estructurada para el inmueble {href_inmueble}')
```

```python
# Se crea dataframe con los datos del primer inmueble con el fin de determinar su estructura
df_inmuebles = extraeInfoInmueble(href_inmueble.iloc[0].href)

# Se recorre desde el segundo hasta el √∫ltimo elemento de la lista del url mediante la
# pasando por la funci√≥n "extraeInfoInmueble" i se guarda la informaci√≥n en el dataframe final
for i in range(1,len(refs)):    
    df_inmuebles = pd.concat([df_inmuebles,extraeInfoInmueble(refs[i])])    
    time.sleep(random.randint(4,8))
```
